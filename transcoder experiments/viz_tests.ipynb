```python
import plotly.express as px
import pandas as pd
import numpy as np
import random
import lorem  # For generating dummy text. Install with: pip install lorem

# Generate dummy data
num_points = 12000
data = {
    "x": np.random.randn(num_points),
    "y": np.random.randn(num_points),
    "text": [lorem.paragraph()[:random.randint(100, 200)] for _ in range(num_points)]
}

df = pd.DataFrame(data)

# Create interactive scatter plot
fig = px.scatter(
    df,
    x="x",
    y="y",
    hover_data={"text": True},
    title="Interactive Scatterplot with Text on Hover"
)

# Optional: Clean up hover template
fig.update_traces(
    hovertemplate="<b>Text:</b><br>%{customdata[0]}<extra></extra>",
    customdata=df[['text']].values
)

# Show plot
fig.show()
```
